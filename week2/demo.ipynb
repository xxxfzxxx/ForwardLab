```python
is_NaN = df.isnull()
row_has_NaN = is_NaN.any(axis=1)
rows_with_NaN = df[row_has_NaN]

rows_with_NaN
```

```python
df.fillna('unknown author', inplace=True)
```

```python
for i in range(len(df)):
   df['references'][i] =  df['references'][i][1:-1]
df.head()
```

Create Schema

```python
from whoosh.fields import Schema, TEXT, KEYWORD, ID, STORED, NUMERIC
from whoosh.analysis import StemmingAnalyzer
schema = Schema(
                index=ID(unique=True),
                title=TEXT(analyzer=StemmingAnalyzer(),stored=True), 
                authors = KEYWORD,
                year = NUMERIC,
                venue=TEXT,  
                abstract = TEXT(analyzer=StemmingAnalyzer()),
                references=TEXT
)

import os, os.path
from whoosh import index

if not os.path.exists("indexdir10"):
    os.mkdir("indexdir10")

idx = index.create_in("indexdir10", schema)
```

```python
idx = index.open_dir("indexdir10")
```

```python
writer=idx.writer()
for i in range(len(df)):
    writer.add_document(
                        index=df.loc[i]['index'].astype('unicode'),
                        title=df.loc[i]['title'],
                        authors=df.loc[i]['authors'],
                        year=df.loc[i]['year'],
                        venue=df.loc[i]['publication venue'],
                        abstract=df.loc[i]['abstract'],
                        references = df.loc[i]['references']
    )
writer.commit()
```

Keyword Search

```python
def search(field, keyword):
    from whoosh.qparser import QueryParser
    from whoosh import scoring
    with idx.searcher(weighting=scoring.TF_IDF()) as searcher:
        query = QueryParser(field, idx.schema).parse(keyword)
        results = searcher.search(query)
        for r in results:
            print(r, r.score)
```

```python
search('venue', 'MIT')
```

```python
def multifield_search(field_list, keyword):
    from whoosh.qparser import MultifieldParser
    from whoosh import scoring
    with idx.searcher(weighting=scoring.TF_IDF()) as searcher:
        query = MultifieldParser(field_list, idx.schema).parse(keyword)
        results = searcher.search(query)
        for r in results:
            print(r, r.score)
```

```python
multifield_search(['title'], 'database management system')
```
